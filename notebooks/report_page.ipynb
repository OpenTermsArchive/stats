# CGUs Statistics

```python
import json
import pandas as pd
import numpy as np
```

```python
with open("../reports/stats_20201029_161524.json") as f:
    data = json.loads(f.read())
```

```python
df = pd.DataFrame.from_dict(data, orient="index")
df
```

## Stats by CGU types

```python
df.groupby("cgu_type").agg(num_docs = ("num_words", "size"), mean_num_words = ("num_words", np.mean)).sort_values("num_docs", ascending=False)
```
