# CGUs Statistics

```python
import json
import pandas as pd
pd.set_option('display.max_rows', None)
import numpy as np
```

```python
with open("../reports/stats_20201105_091749.json") as f:
    data = json.loads(f.read())
```

```python
df = pd.DataFrame.from_dict(data, orient="index")
df
```

## Stats by CGU types

```python
df.groupby("document_type").agg(num_docs = ("num_words", "size"), mean_num_words = ("num_words", np.mean), ecart_type = ("num_words", np.std)).sort_values("num_docs", ascending=False)
```
